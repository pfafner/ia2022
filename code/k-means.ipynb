```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import k_means

from sklearn.datasets import load_iris, make_circles, make_moons, make_blobs
from skimage.data import chelsea, coffee
```

# K-means

```python
data = load_iris()
```

```python
X = data['data']
y = data['target']
```

```python
plt.figure()
plt.scatter(X[:,0], X[:,1], c=y)
plt.show()
```

```python
k = 3
centers, labels, u = k_means(X, n_clusters=k, algorithm='auto', init='random')
```

```python
labels
```

```python
plt.figure()
plt.scatter(X[:,0], X[:,1], c=labels)
plt.show()
```

```python
plt.figure()
plt.plot(centers[:,0], centers[:,1], 'o')
plt.show()
```

## Image Quantization

```python
I = plt.imread('horse.jpg')
```

```python
plt.figure()
plt.imshow(I)
plt.show()
```

```python
sh = I.shape
sh
```

```python
I[0,0,:]
```

```python
# matriz de datos
Iflat = I.reshape(-1,3)
```

```python
Iflat.shape
```

```python
Iflat[:10]
```

```python
k = 2     # nmero de clsters
centers, labels, u = k_means(Iflat, n_clusters=k, algorithm='auto', init='random')
```

```python
plt.figure()
plt.imshow(labels.reshape(sh[0], sh[1]), cmap='seismic')
plt.show()
```

```python
result = np.zeros(sh).reshape(-1,3)

# asignamos a cada pixel su centroide (color promedio)
for i in range(0, k):
    result[labels == i] = centers[i] 

result = result.reshape(sh).astype(np.uint8)
```

```python
plt.figure()
plt.imshow(result)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(I)
plt.subplot(1,2,2)
plt.imshow(result)
plt.show()
```
